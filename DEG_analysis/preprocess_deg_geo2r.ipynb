```python
import pandas as pd
import numpy as np

# ----------------------------------------
# Load GEO2R output (TSV)
# ----------------------------------------
deg = pd.read_csv("GEO2R_full_table.tsv", sep="\t")

# Remove probes without gene symbols
deg = deg[deg["Gene.symbol"].notna()].copy()

# Remove ambiguous probes (///)
deg = deg[~deg["Gene.symbol"].str.contains("///")].copy()

# Convert numeric columns
deg["adj.P.Val"] = pd.to_numeric(deg["adj.P.Val"], errors="coerce")
deg["logFC"] = pd.to_numeric(deg["logFC"], errors="coerce")

# ----------------------------------------
# DEG filtering criteria
# ----------------------------------------
p_thresh = 0.01
fc_thresh = 0.58

deg_filt = deg[
    (deg["adj.P.Val"] < p_thresh) &
    (deg["logFC"].abs() > fc_thresh)
].copy()

# Volcano plot helper
deg_filt["neglog10p"] = -np.log10(deg_filt["adj.P.Val"])

# Save output
deg_filt.to_csv("DEG_filtered_clean.tsv", sep="\t", index=False)
```
